```python
# 포켓몬 이름 번역기
# https://wiki.xn--rckteqa2e.com/wiki/%E3%83%9D%E3%82%B1%E3%83%A2%E3%83%B3%E3%81%AE%E5%A4%96%E5%9B%BD%E8%AA%9E%E5%90%8D%E4%B8%80%E8%A6%A7#%E9%96%A2%E9%80%A3%E9%A0%85%E7%9B%AE
# 여기에 있는 포켓몬 이름을 가져오자.
# 우선 일본어, 영어, 한국어만

# 이름 데이터 가져오기
import sys
import requests
import json
import re

url = 'https://wiki.xn--rckteqa2e.com/wiki/%E3%83%9D%E3%82%B1%E3%83%A2%E3%83%B3%E3%81%AE%E5%A4%96%E5%9B%BD%E8%AA%9E%E5%90%8D%E4%B8%80%E8%A6%A7#%E9%96%A2%E9%80%A3%E9%A0%85%E7%9B%AE'
res = requests.get(url)
soup = BeautifulSoup(res.text, 'lxml')

tables = soup.select('table', class_='graytable c')
json_data = []
pattern = r'^([^()]+)'

for table in tables:
    trs = table.select('tr')
    for tr in trs:
        tds = tr.select('td')
        json_item = {}
        if tds:
            json_item['num'] = tds[0].get_text().replace('\n','')
            json_item['jpn'] = tds[1].get_text().replace('\n','')
            
            # 영어열은 특별처리
            eng_ori = tds[2].get_text().replace('\n','')
            match = re.search(pattern, eng_ori)
            if match:
                eng = match.group(1).strip()
            else:
                eng = eng_ori
            
            json_item['eng'] = eng
            
            json_item['deu'] = tds[3].get_text().replace('\n','')
            json_item['fra'] = tds[4].get_text().replace('\n','')
            json_item['kor'] = tds[5].get_text().replace('\n','')
            json_item['chn'] = tds[6].get_text().replace('\n','')
            json_item['twn'] = tds[7].get_text().replace('\n','')
            
            json_data.append(json_item)
    
json_file_name = 'pokemon_name_multilan.json'

with open(json_file_name, 'w', encoding='utf-8') as f:
    json.dump(json_data, f, ensure_ascii=False, indent =4)
    
# 번역 함수
def translate_pokemon(name,ori_lan,target_lan):
    LANGUAGE_LIST = ['jpn','eng','deu','fra','kor','chn','twn']
    
    if ori_lan in LANGUAGE_LIST and target_lan in LANGUAGE_LIST:
        json_file_name = 'pokemon_name_multilan.json'
        with open(json_file_name, 'r', encoding='utf-8') as f:
            pokemon_data = json.load(f)
        
        for pokemon in pokemon_data:
            if pokemon[ori_lan].strip().lower() == name.strip().lower():
                return pokemon[target_lan].strip()
            
        print("Please check the name of pokemon")
        return False
    else:
        print("Please input valid original and target language")
        print("Languages : ['jpn','eng','deu','fra','kor','chn','twn']")
        return False
```

```python
# 이미지 합성
# 좌측에 카드이미지
# 우측에 이름과 타입, 효과가 나오도록
# 일단 테라파고스로 테스트

from PIL import Image, ImageDraw, ImageFont
from IPython.display import display
import textwrap
    
def add_text_to_image(original_image_path, text_list, output_path, font_path, font_size):
    # 원본 이미지 열기
    original_image = Image.open(original_image_path)
    
    # 원본 이미지의 너비와 높이 가져오기
    original_width, original_height = original_image.size
    
    # 새로운 이미지 생성하기 (원본 이미지의 절반 크기)
    new_width = original_width * 2  # 가로는 원본 이미지의 2배
    new_height = original_height    # 세로는 원본 이미지와 동일
    new_image = Image.new('RGB', (new_width, new_height), color='white')
    
    # 새 이미지에 원본 이미지 붙여넣기 (왼쪽에 위치)
    new_image.paste(original_image, (0, 0))
    
    # 텍스트 리스트를 \n 으로 이어주기
    text = ''
    for text_item in text_list:
        text += text_item + '\n'
    
    # 새 이미지에 텍스트 추가하기 (오른쪽에 위치)
    draw = ImageDraw.Draw(new_image)
    font = ImageFont.truetype(font_path, font_size)
    text_x = original_width 
    text_y = (new_height) // 2
    draw.text((text_x, text_y), text, fill="black", font=font)
    
    # 새로운 이미지 저장
    new_image.save(output_path)
    display(new_image)

# 예시 사용
item =    {
        "id": 88,
        "type": "포켓몬",
        "name_jp": "テラパゴスex",
        "name_kr": "테라파고스ex",
        "image_url": "https://www.pokemon-card.com/assets/images/card_images/large/SV7/046027_P_TERAPAGOSUEX.jpg",
        "image_path": "./stella_miracle/images/88.jpg",
        "text": ["[기본]",
            "무무 유니온비트 30×",
            "이 기술은 후공 플레이어의 첫턴에는 사용할 수 없다. 자신의 벤치 포켓몬의 수 × 30 데미지를 준다.",
            "풀물번 크라운오팔 180",
            "다음 상대의 턴 이 포켓몬은 기본 포켓몬(무색 포켓몬은 제외)으로부터 기술의 데미지를 받지않는다."]
}

font_path = '/Library/Fonts/Arial Unicode.ttf'  # 사용하려는 폰트의 경로
font_size = 40  # 폰트 크기

add_text_to_image(item['image_path'], item['text'], './stella_miracle/translated/' + str(item['id']) + '.jpg', font_path, font_size)

```

```python
from PIL import Image, ImageDraw, ImageFont
from IPython.display import display
import textwrap

def add_text_to_image(original_image_path, text_list, name, output_path, font_path, font_size):
    # 원본 이미지 열기
    original_image = Image.open(original_image_path)
    
    # 원본 이미지의 너비와 높이 가져오기
    original_width, original_height = original_image.size
    
    # 새로운 이미지 생성하기 (원본 이미지의 절반 크기)
    new_width = original_width * 2  # 가로는 원본 이미지의 2배
    new_height = original_height    # 세로는 원본 이미지와 동일
    new_image = Image.new('RGB', (new_width, new_height), color='white')
    
    # 새 이미지에 원본 이미지 붙여넣기 (왼쪽에 위치)
    new_image.paste(original_image, (0, 0))
    
    # 텍스트 리스트를 줄바꿈 기호로 이어주기
    text = '\n'.join(text_list)
    
    # 새 이미지에 텍스트 추가하기 (오른쪽에 위치)
    draw = ImageDraw.Draw(new_image)
    font = ImageFont.truetype(font_path, font_size)
    
    # 텍스트 줄 나누기
    max_text_width = original_width - 50  # 여백을 고려하여 텍스트 최대 너비 설정
    lines = []
    for text_item in text_list:
        wrapped_lines = textwrap.wrap(text_item, width=int(max_text_width / (font_size * 0.85)))  # 대략적인 문자 당 너비로 나누기
        lines.extend(wrapped_lines)
        
    # 각 줄의 높이 계산
    # line_height = int(font.getlength('A')[1])
    line_height = 50
    total_text_height =line_height * len(lines)
    
    # 텍스트 위치 계산
    text_x = original_width + 40# 오른쪽 여백
    text_y = (new_height - total_text_height) / 2 # 중앙 정렬
    
    # 각 줄을 그리기
    for line in lines:
        draw.text((text_x, text_y), line, fill="black", font=font)
        text_y += line_height * 1.5
        
    # 카드 이름 그리기
    name_x = original_width + 40
    name_y = 40
    name_font = ImageFont.truetype(font_path, font_size*2)
    draw.text((name_x,name_y), name,fill="black", font=name_font)
    
    # 가로줄 그리기
    draw.line([(original_width+40, 150), (new_width-40, 150)], fill="black", width=5)
    
    
    # 새로운 이미지 저장
    new_image.save(output_path)
    display(new_image)

# 예시 사용
item = {
    "id": 88,
    "type": "포켓몬",
    "name_jp": "テラパゴスex",
    "name_kr": "테라파고스ex",
    "image_url": "https://www.pokemon-card.com/assets/images/card_images/large/SV7/046027_P_TERAPAGOSUEX.jpg",
    "image_path": "./stella_miracle/images/88.jpg",
    "text": [
        "[기본]",
        "무무 유니온비트 30×",
        "이 기술은 후공 플레이어의 첫턴에는 사용할 수 없다. 자신의 벤치 포켓몬의 수 × 30 데미지를 준다.",
        "풀물번 크라운오팔 180",
        "다음 상대의 턴 이 포켓몬은 기본 포켓몬(무색 포켓몬은 제외)으로부터 기술의 데미지를 받지않는다."
    ]
}

font_path = '/Library/Fonts/Arial Unicode.ttf'  # 사용하려는 폰트의 경로
font_size = 40  # 폰트 크기

add_text_to_image(item['image_path'], item['text'], item['name_kr'],'./stella_miracle/translated/' + str(item['id']) + '.jpg', font_path, font_size)
```
