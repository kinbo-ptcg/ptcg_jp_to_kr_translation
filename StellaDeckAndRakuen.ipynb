```python
# 스텔라덱 이미지,정보 겟
# 스텔라덱의 카드아이디는 46184~ 46205(파라블레이즈), 46206~ 46227(님피아)
# url : https://www.pokemon-card.com/card-search/details.php/card/{id}
# id =  46184~46227

from bs4 import BeautifulSoup
import sys
import requests
import os
import json

ID_FIRST = 46184
ID_LAST = 46227
url_head = 'https://www.pokemon-card.com/card-search/details.php/card/'
url_image_head = 'https://www.pokemon-card.com'
TYPES_NONPOKEMON = ['グッズ', 'ポケモンのどうぐ','サポート','スタジアム','トレーナー']

json_data = []

for id_card in range(ID_FIRST, ID_LAST+1):
    url = url_head + str(id_card)
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'lxml')
    
    card_name = soup.select('.Heading1.mt20')[0].get_text()
    card_type = soup.select('.RightBox .mt20')[0].get_text()
    if card_type not in TYPES_NONPOKEMON:
        card_type = 'ポケモン'
        
    json_item = {}
    json_item['id'] = id_card - ID_FIRST + 1
    json_item['name_jp'] = card_name
    json_item['type_jp'] = card_type
    json_item['image_url'] = url_image_head + soup.select('.LeftBox .fit')[0].get('src')
    
    json_data.append(json_item)
    
    url_image = soup.select('.LeftBox .fit')[0].get('src')
    url_image = url_image_head + url_image
    print(url_image)
    
    # get image of card
    image_res = requests.get(url_image)
    image_dir = "./stedeck_dragona/images/"
    image_name = "stedeck" + str(id_card-ID_FIRST+1) +".jpg"
    image_path = image_dir + image_name
    
    # 저장할 디렉토리가 없으면 생성
    if not os.path.exists(image_dir):
        os.makedirs(image_dir)
        
    with open(image_path,'wb') as file:
        file.write(image_res.content)

json_file_name = 'stella_deck_data.json'

with open(json_file_name, 'w', encoding='utf-8') as f:
    json.dump(json_data, f, ensure_ascii=False, indent =4)
```

```python
# 한글판 카드 간이 이미지

from PIL import Image, ImageDraw, ImageFont
from IPython.display import display
import textwrap

def add_text_to_image(original_image_path, text_list, name, output_path, font_path, font_size):
    # 원본 이미지 열기
    original_image = Image.open(original_image_path)
    
    # 원본 이미지의 너비와 높이 가져오기
    original_width, original_height = original_image.size
    
    # 새로운 이미지 생성하기 (원본 이미지의 절반 크기)
    new_width = original_width * 2  # 가로는 원본 이미지의 2배
    new_height = original_height    # 세로는 원본 이미지와 동일
    new_image = Image.new('RGB', (new_width, new_height), color='white')
    
    # 새 이미지에 원본 이미지 붙여넣기 (왼쪽에 위치)
    new_image.paste(original_image, (0, 0))
    
    # 텍스트 리스트를 줄바꿈 기호로 이어주기
    text = '\n'.join(text_list)
    
    # 새 이미지에 텍스트 추가하기 (오른쪽에 위치)
    draw = ImageDraw.Draw(new_image)
    font = ImageFont.truetype(font_path, font_size)
    
    # 텍스트 줄 나누기
    max_text_width = original_width - 50  # 여백을 고려하여 텍스트 최대 너비 설정
    lines = []
    for text_item in text_list:
        wrapped_lines = textwrap.wrap(text_item, width=int(max_text_width / (font_size * 0.85)))  # 대략적인 문자 당 너비로 나누기
        lines.extend(wrapped_lines)
        
    # 각 줄의 높이 계산
    # line_height = int(font.getlength('A')[1])
    line_height = 50
    total_text_height =line_height * len(lines)
    
    # 텍스트 위치 계산
    text_x = original_width + 40# 오른쪽 여백
    text_y = (new_height - total_text_height) / 2 # 중앙 정렬
    
    # 각 줄을 그리기
    for line in lines:
        draw.text((text_x, text_y), line, fill="black", font=font)
        text_y += line_height * 1.5
        
    # 카드 이름 그리기
    name_x = original_width + 40
    name_y = 40
    name_font = ImageFont.truetype(font_path, font_size*2)
    draw.text((name_x,name_y), name,fill="black", font=name_font)
    
    # 가로줄 그리기
    draw.line([(original_width+40, 150), (new_width-40, 150)], fill="black", width=5)
    
    
    # 새로운 이미지 저장
    new_image.save(output_path)
    display(new_image)

# 예시 사용
item = {
    "id": 88,
    "type": "포켓몬",
    "name_jp": "テラパゴスex",
    "name_kr": "테라파고스ex",
    "image_url": "https://www.pokemon-card.com/assets/images/card_images/large/SV7/046027_P_TERAPAGOSUEX.jpg",
    "image_path": "./stella_miracle/images/88.jpg",
    "text": [
        "[기본]",
        "(무)(무) 유니온비트 30×",
        "이 기술은 후공 플레이어의 첫턴에는 사용할 수 없다. 자신의 벤치 포켓몬의 수 × 30 데미지를 준다.",
        "(풀)(물)(번) 크라운오팔 180",
        "다음 상대의 턴 이 포켓몬은 기본 포켓몬((무) 포켓몬은 제외)으로부터 기술의 데미지를 받지않는다."
    ]
}

font_path = '/Library/Fonts/Arial Unicode.ttf'  # 사용하려는 폰트의 경로
font_size = 40  # 폰트 크기

add_text_to_image(item['image_path'], item['text'], item['name_kr'],'./' + str(item['id']) + '.jpg', font_path, font_size)
```

```python
import requests
from PIL import Image, ImageDraw, ImageFont
from io import BytesIO
import textwrap
import os
import re

symbol_map = {
    "(풀)": './assets/풀.webp',
    "(불)": './assets/불.webp',
    "(물)": './assets/물.webp',
    "(번)": './assets/번.webp',
    "(초)": './assets/초.webp',
    "(격)": './assets/격.webp',
    "(악)": './assets/악.webp',
    "(강)": './assets/강.webp',
    "(무)": './assets/무.webp',
    "(드)": './assets/드.webp'
}
    
# 패턴을 검출하는 정규표현식 생성
symbol_pattern = re.compile(r'\(풀\)|\(불\)|\(물\)|\(번\)|\(초\)|\(격\)|\(악\)|\(강\)|\(무\)|\(드\)')

def add_text_and_symbols_to_image(original_image_path, text_list, name, output_path, font_path, font_size):
    # 원본 이미지 열기
    original_image = Image.open(original_image_path)
    
    # 원본 이미지의 너비와 높이 가져오기
    original_width, original_height = original_image.size
    
    # 새로운 이미지 생성하기 (원본 이미지의 2배 너비)
    new_width = original_width * 2  
    new_height = original_height
    new_image = Image.new('RGB', (new_width, new_height), color='white')
    
    # 새 이미지에 원본 이미지 붙여넣기 (왼쪽)
    new_image.paste(original_image, (0, 0))
    
    # 텍스트 줄바꿈 처리 및 추가
    draw = ImageDraw.Draw(new_image)
    font = ImageFont.truetype(font_path, font_size)
    
    # 텍스트 줄 나누기
    max_text_width = original_width - 50  # 여백을 고려한 최대 텍스트 너비
    lines = []
    for text_item in text_list:
        wrapped_lines = textwrap.wrap(text_item, width=int(max_text_width / (font_size * 0.85)))
        lines.extend(wrapped_lines)
    
    # 텍스트 높이 및 위치 계산
    line_height = 50
    total_text_height = line_height * len(lines)
    
    text_x = original_width + 40
    text_y = (new_height - total_text_height) / 2
    
    # 각 줄을 그리기 (특정 기호는 이미지로 대체)
    for line in lines:
        # line 내에서 symbol_map에 해당하는 패턴 찾기
        while line:
            match = symbol_pattern.search(line)
            if match:
                # 패턴이 등장한 위치 이전 텍스트 출력
                pre_text = line[:match.start()]
                draw.text((text_x, text_y), pre_text, fill="black", font=font)
                text_x += draw.textsize(pre_text, font=font)[0] + 10
                
                # 패턴을 이미지로 대체
                symbol = match.group(0)
                symbol_image = Image.open(symbol_map[symbol])
                symbol_size = (int(font_size * 1.5), int(font_size * 1.5))  # 심볼 크기 조정
                symbol_image = symbol_image.resize(symbol_size)
                new_image.paste(symbol_image, (int(text_x), int(text_y)), symbol_image)
                text_x += symbol_size[0] - 10  # 이미지 너비만큼 X 좌표 이동
                
                # line에서 처리된 부분 잘라내기
                line = line[match.end():]
            else:
                # 더 이상 패턴이 없을 때 나머지 텍스트 출력
                draw.text((text_x, text_y), line, fill="black", font=font)
                text_x += draw.textsize(line, font=font)[0] + 10
                break
        
        # 줄 간격 추가
        text_x = original_width + 40  # 다시 처음 X 좌표로 이동
        text_y += line_height * 1.5
    
    # 카드 이름 그리기
    name_font = ImageFont.truetype(font_path, font_size * 2)
    draw.text((original_width + 40, 40), name, fill="black", font=name_font)
    
    # 가로줄 그리기
    draw.line([(original_width + 40, 150), (new_width - 40, 150)], fill="black", width=5)
    
    # 이미지 저장
    new_image.save(output_path)
    print(f"Image saved to {output_path}")
    #display(new_image)

# 사용 예시
item = {
    "id": 88,
    "type": "포켓몬",
    "name_jp": "テラパゴスex",
    "name_kr": "테라파고스ex",
    "image_url": "https://www.pokemon-card.com/assets/images/card_images/large/SV7/046027_P_TERAPAGOSUEX.jpg",
    "image_path": "./stella_miracle/images/88.jpg",
    "text": [
        "[기본]",
        "(무)(무) 유니온비트 30×",
        "이 기술은 후공 플레이어의 첫턴에는 사용할 수 없다. 자신의 벤치 포켓몬의 수 × 30 데미지를 준다.",
        "(풀)(물)(번) 크라운오팔 180",
        "다음 상대의 턴 이 포켓몬은 기본 포켓몬((무) 포켓몬은 제외)으로부터 기술의 데미지를 받지않는다."
    ]
}

font_path = '/Library/Fonts/Arial Unicode.ttf'  # 사용하려는 폰트의 경로
font_size = 40  # 폰트 크기

# 이미지에 텍스트 추가
#add_text_and_symbols_to_image(item['image_path'], item['text'], item['name_kr'], f"./{item['id']}.jpg", font_path, font_size)
```
